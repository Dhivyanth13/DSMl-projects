```python
df = df.drop(columns=["Address", "Locality Verbose","Locality","Longitude","Latitude","Currency","Restaurant ID",
                      "Restaurant Name","Rating color","Rating text"])
```

```python
df["Has Table booking"]=df["Has Table booking"].map({"Yes": 1, "No": 0})
df["Has Online delivery"]=df["Has Online delivery"].map({"Yes": 1, "No": 0})
df["Is delivering now"]=df["Is delivering now"].map({"Yes": 1, "No": 0})
df["Switch to order menu"]=df["Switch to order menu"].map({"Yes": 1, "No": 0})
```

```python
df['City_freq'] = df['City'].map(df['City'].value_counts())
```

```python
df=df.drop(columns=["City"])
```

```python
# Example mapping dictionary
cuisine_map = {
    'North Indian': 'Indian',
    'South Indian': 'Indian',
    'Mughlai': 'Indian',
    'Chinese': 'Asian',
    'Thai': 'Asian',
    'Japanese': 'Asian',
    'Italian': 'European',
    'French': 'European',
    'Mexican': 'Latin American',
    'American': 'Western',
    'Continental': 'Western'
}

# Apply mapping
# Apply mapping to the correct column
df['Cuisine_grouped'] = df['Cuisines'].map(cuisine_map).fillna('Other')

# Encode grouped target
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Cuisine_encoded'] = le.fit_transform(df['Cuisine_grouped'])
```

```python
df
```

```python
df=df.drop(columns=["Cuisines","Cuisine_grouped"])
```

```python
df
```

```python
x=df.drop(columns=["Cuisine_encoded"])
y=df["Cuisine_encoded"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
