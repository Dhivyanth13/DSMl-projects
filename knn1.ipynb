```python
import pandas as pd
```

```python
i=pd.read_csv("iphone_purchase_records.csv")
i.head()
```

```python
i.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
i['Gender']=l.fit_transform(i['Gender'])
i['Gender']
```

```python
#feature engineering
x=i[['Gender','Age','Salary']]
y=i['Purchase Iphone']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sso=StandardScaler()
sx_train=sso.fit_transform(x_train)
sx_train
```

```python
sx_test=sso.transform(x_test)
sx_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=KNeighborsClassifier(n_neighbors=3,metric='euclidean')
k.fit(sx_train,y_train)
```

```python
y_pred=k.predict(sx_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
