```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv(r"C:\Users\dhivy\Downloads\archive (1)\india_housing_prices.csv")
plt.figure(figsize=(8,5))
sb.histplot(df['Price_in_Lakhs'],bins=40,kde=True)
plt.title("Distribution of house prices")
plt.show
```

```python
df.columns
```

```python
sb.countplot(x='BHK',data=df)
plt.title("Distribution of BHK")
```

```python
plt.figure(figsize=(8,5))
sb.kdeplot(
    x=df['Size_in_SqFt'],
    y=df['Price_in_Lakhs'],
    fill=True,
    cmap='Blues',
    thresh=0.05
)
plt.title("Density of Size vs Price")
plt.xlabel("Size in SqFt")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sb.boxplot(x="Property_Type",y='Price_in_Lakhs',data=df)
plt.xticks(rotation=90)
plt.title("price distribution by prop type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sb.heatmap(
    df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm',fmt=".2f")
plt.title("correlation matrix")
plt.show()
```

```python
df['Price_Category']=pd.cut(df['Price_in_Lakhs'],bins=[0,100,250,400,600],labels=['low','mid','high','luxury'])
sb.countplot(x='Price_Category',hue='Property_Type',data=df)
```
