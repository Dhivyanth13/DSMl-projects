```python
import pandas as pd
```

```python
a=pd.read_csv("insurance.csv")
a
```

```python
a= pd.get_dummies(a, columns=['region'], drop_first=True)
```

```python
a.isnull().sum()
```

```python
a['sex'] = a['sex'].map({'male': 1, 'female': 0})
a['smoker'] = a['smoker'].map({'yes': 1, 'no': 0})
```

```python
a.describe()
```

```python
a.info()
```

```python
a.head()
```

```python
x=a[['age','bmi','children','region_northwest','region_southeast','region_southwest','smoker','sex']]
y=a['charges']
x
```

```python
#model selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
x_test
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
#accuracy measure
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
