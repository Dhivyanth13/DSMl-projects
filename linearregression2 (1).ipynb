```python
import pandas as pd
```

```python
s=pd.read_csv("dataset_study.csv")
s
```

```python
s=s.drop("student_id",axis="columns")
s
```

```python
s.isnull().sum()
```

```python
#data exploration
s.info()
```

```python
s.describe()
```

```python
#feature engineering
import numpy as np
x=s[["study_hours"]]
y=s["grade"]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
k=StandardScaler()
kx_train=k.fit_transform(x_train)
kx_train
```

```python
kx_test=k.transform(x_test)
kx_test
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(kx_train, y_train)

print("R²:", ridge.score(kx_test, y_test))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
